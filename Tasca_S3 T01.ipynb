### Exercici 1

*Crea una funció que donat un Array d’una dimensió, et faci un resum estadístic bàsic de les dades. Si detecta que l’array té més d’una dimensió, ha de mostrar un missatge d’error.*

```python
def stat_basic(arr):
    if arr.ndim <= 1:
        print(f'media aritmetica: {np.mean(arr)}')
        print(f'mediana: {np.median(arr)}')
        print(f'desviación tipica: {np.std(arr)}')
        print(f'varianza: {np.var(arr)}')
    else:
        print('ERROR')
```

```python
arr1 = np.array([3,5,7])

stat_basic(arr1)
```

### Exercici 3

Crea una funció que donada una taula de dues dimensions (NxM), et calculi els totals per fila i els totals per columna.

```python
def matriu_totals():
    N = int(input("Enter número files: "))
    M = int(input("Enter número columnes: "))
    
    array = np.random.randint(10, size=(N,M)) #valors aleatoris entre el 0 i el 10.
    
    print(array)
    
    return np.sum(array, axis=1), np.sum(array, axis=0) #calcul totals fila i columna
```

### Exercici 4

Implementa manualment una funció que calculi el coeficient de correlació. Informa-te'n sobre els seus usos i interpretació.

* El coeficiente de correlación de la muestra puede representarse con la siguiente fórmula:

* Vamos a ver cómo calcular el coeficiente de correlación a través de un ejemplo entre dos variables (altura y peso).

```python
#introducción de datos
altura = np.array([1.75, 1.59, 1.66, 1.66, 1.68])
peso = np.array([82, 68, 70, 61, 70])

#definición de variables
n = len(altura) #Número total de elementos
x_mean = np.mean(altura) #Calcular la media de X
y_mean = np.mean(peso) #Calcular la media de Y
st_dev_x = np.std(altura) #calcular la desviación estandard de X
st_dev_y = np.std(peso) #Calcular la desviación estandard de Y
```

```python
#Función para calcular desviación de cada punto "X" e "Y" respecto a la media 

def num1(altura):
    list = []

    for n in altura:
        var1 = n - x_mean
        list.append(var1)
    
    return list

def num2(peso):
    list1 = []
    
    for p in peso:
        var2 = p - y_mean
        list1.append(var2)
    
    return list1
```

```python
#Completamos el numerador de la ecuación
var3 = sum(np.multiply(num1(altura),num2(peso)))
print(var3)
```

```python
#Hallamos el coeficiente de correlación
r = (1/n)*(var3)/(st_dev_x*st_dev_y)
print(round(r, 8))
```

```python
#Comparamos el resultado obtenido con el de la formula
np.corrcoef(altura, peso)
```

Existeix una correlació positiva, on alçada i pes estan fortament relacionats
