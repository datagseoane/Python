Crea almenys una visualització per:

* Una variable categòrica (Districte o Barri)

```python
# Preparar los datos
plotdata = df.Nom_Districte.value_counts()

#Gráfico de barras + Títulos
plotdata.plot(kind='bar', title = 'Registres per a cada districte')
plt.xlabel("Nom Districte")
plt.ylabel("Registres")

# Guardar imagen
plt.savefig("1_Districte.png")
```

```python
#alternativa en seaborn
sns.countplot(x=df.Nom_Districte)
```

* Una variable numèrica (Preu)

```python
# Preparar los datos
df.Preu = df.Preu.replace('--','0')
df.Preu = df.Preu.astype(float)
df.info()
```

```python
data1 = df.Preu
data1.plot(kind='hist')

# Guardar imagen
plt.savefig("2_PreuLloguer.png")
```

```python
#alternativa en seaborn
sns.histplot(x=df.Preu, bins=10)
```

* Una variable numèrica i una categòrica (Districte i Preu)

```python
# Preparar los datos
data2 = df.groupby('Nom_Districte').Preu.mean().round(1)

#Gráfico de barras + Títulos
data2.plot(kind='bar', title = 'Lloguer mitjà mensual per a cada districte')
plt.xlabel("Nom Districte")
plt.ylabel("Euros")

# Guardar imagen
plt.savefig("3_Districte_preu.png")
```

* Dues variables numèriques (Any o Trimestre i Preu)

```python
# Preparar los datos
data3 = df.groupby('Trimestre').Preu.mean().round(1)

#Gráfico de barras + Títulos
data3.plot(kind='bar', title = 'Lloguer mitjà mensual per Trimestre')
plt.xlabel("Trimestre")
plt.ylabel("Euros")

# Guardar imagen
plt.savefig("4_Trimestre_preu.png")
```

```python
sns.scatterplot(x="Trimestre", y="Preu", data=df)
```

* Tres variables (Barri o Districte, Trimestre i Preu)

```python
# Preparar los datos
data4 = df.groupby(['Nom_Districte','Trimestre']).Preu.mean().round(1).unstack()
print(data4)
```

```python
#Gráfico de barras + Títulos
data4.plot(kind='bar', title = 'Lloguer mitjà mensual per a cada Districte')
plt.xlabel("Nom Districte")
plt.ylabel("Euros")

# Guardar imagen
plt.savefig("5_evolucion_por_barrios.png")
```
