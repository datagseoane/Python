Crea almenys una visualització per:

* Una variable categòrica (Districte o Barri)

```python
plotdata = df.Nom_Districte.value_counts()
plotdata.plot(kind='bar', title = 'Registres per a cada districte')
plt.xlabel("Nom Districte")
plt.ylabel("Registres")
plt.savefig("Districte.png")
```

* Una variable numèrica (Preu)

```python
df.Preu = df.Preu.replace('--','0')
df.Preu = df.Preu.astype(float)
df.info()
```

```python
data1 = df.Preu
data1.plot(kind='hist')
plt.savefig("PreuLloguer.png")
```

* Una variable numèrica i una categòrica (Districte i Preu)

```python
data2 = df.groupby('Nom_Districte').Preu.mean().round(1)
data2.plot(kind='bar', title = 'Lloguer mitjà mensual per a cada districte')
plt.xlabel("Nom Districte")
plt.ylabel("Euros")
plt.savefig("districte_preu.png")
```

* Dues variables numèriques (Any o Trimestre i Preu)

```python
data3 = df.groupby('Trimestre').Preu.mean().round(1)
data3.plot(kind='bar', title = 'Lloguer mitjà mensual per a cada districte')
plt.xlabel("Nom Districte")
plt.ylabel("Euros")
plt.savefig("Trimestre_preu.png")
```

* Tres variables (Barri o Districte, Trimestre i Preu)

```python
data4 = df.groupby(['Nom_Districte','Trimestre']).Preu.mean().round(1).unstack()
print(data4)
```

```python
data4.plot(kind='bar')
plt.xlabel("Nom Districte")
plt.ylabel("Euros")

plt.savefig("evolucion_por_barrios.png")
```
